# BIVARIATE ANALYSIS

```python
pd.crosstab(titanic['Survived'],titanic['Pclass'],normalize="columns")*100
```

```python
# Heatmap is a type of visualization technique used to show differences in data through variation in colours.
group = titanic.groupby(['Pclass', 'Survived'])
pclass_survived = group.size().unstack()
sns.heatmap(pclass_survived, annot = True,cmap="plasma",linewidths=1, fmt ="d")
plt.title('Heatmap of passangers survived based on Pclass', fontsize=20)
```

#### This heatmap shows the number of passanger survived based on the Pclass of the passanger. From the heatmap, it is clear that the passenger with the Pclass 1 has survived more than the passengers with other Pclass.

```python
pd.crosstab(titanic['Survived'],titanic['Embarked'],normalize="columns")*100
```

```python
sns.countplot(x='Embarked', hue='Survived', data=titanic)
plt.title('Survival Count by Port of Embarkation')
plt.xlabel('Port of Embarkation')
plt.ylabel('Count')
plt.show()
```

#### This shows that the passengers from port cherbourg has survived more and the passenger from queenstown has not survived at all.

```python
pd.crosstab(titanic['Sex'],titanic['Embarked'],normalize="columns")*100
```

```python
sns.violinplot(x='Embarked', y='Age', hue='Sex', data=titanic, split=True)
plt.title('Age Distribution by Port of Embarkation and Gender')
plt.xlabel('Port of Embarkation')
plt.ylabel('Age')
plt.show()
```

#### 1) It concludes that the female and male on port S is mainly of the age between 20-60.
#### 2) The male and female on port C is mainly between 20-40.

```python
pd.crosstab(titanic['Pclass'],titanic['Embarked'],normalize="columns")*100
```

```python
sns.catplot(x='Pclass', hue='Embarked', data=titanic, kind='count')
```

#### 1) According to this heat map, it is concluded that most of the passengers of Pclass 1 has embarkation port as C.
#### 2) No passenger from Pclass 2 and 3 has embarkation port as Q.


```python
# Bar Chart is a easy-to-use method of data visualization
titanic['Age']= pd.qcut(titanic['Age'],4)
sns.barplot(x='Age',y='Survived',
data = titanic)
```

#### Barplot is plotted between the number of passengers survived based on the age group. According to this, it concludes that the passengers between age 27 to 39 has survived more.

```python
# changing datatype from float to int
titanic['Fare'] = titanic['Fare'].astype(int)
print(titanic.dtypes)
```

```python
#line plot for fare
plt.plot(titanic['Fare'], color='blue', marker='o', linestyle='-')
plt.xlabel('Index')
plt.ylabel('Fare')
plt.title('Line Graph of Fare')
plt.show()
```

```python
# changing datatype of age from float to int
titanic['Age'] = titanic['Age'].apply(lambda x: x.mid)
titanic['Age'] = titanic['Age'].astype(int)
print(titanic.dtypes)
```

```python
x = titanic['Age']  
y = titanic['Fare']  
bubble_size = titanic['Pclass'] 
plt.scatter(x, y, s=bubble_size*10, alpha=0.5)  
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Bubble Plot: Age vs. Fare (Bubble size indicates Pclass)')
plt.show()
```
